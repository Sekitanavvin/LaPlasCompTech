```python
import scipy.io as sio
from scipy import signal
import numpy as np
import matplotlib.pyplot as plot
 
mat_fname = "C:\\уеба\\мага\\питончик\\LaPlasCompTech\\LaPlasCompTech\\33484K-HFS.mat"
 
mat_contents = sio.loadmat(mat_fname)

sig = mat_contents['sigKH']
tb = mat_contents['tbKH']
sig1 = np.reshape(sig,(sig.size,))
tb1 = np.reshape(tb,(tb.size,))

def serial_corr(wave, lag=1):
    n = len(wave)
    y1 = wave[lag:]
    y2 = wave[:n-lag]
    corr = np.corrcoef(y1, y2)[0, 1]
    return corr

def autocorr(wave):
    lags = range(len(wave)//10)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

autoc, autolag = autocorr(sig1[5000:105001])
plot.plot(autoc, autolag)
plot.xlabel('Lag')
plot.ylabel('Correlation')
plot.show()
```
